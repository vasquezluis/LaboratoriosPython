## Configuración inicial

```python
# importar las librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# buscar el archivo que necesitamos
path = "/home/luii/Documents/Modelacionysimulacion"
file = "operacionesmayo.csv"
```

```python
# leer el archivo y mostrarlo
data = pd.read_csv(path + "/" + file)
data
```

## Tiendas que han cumplido metas

```python
# Enlistar las tiendas que han cumplido las metas, usando string
tablaMetaSi = data[data["MetaCumplida"] == "Yes"]
tablaMetaSi
```

## Gráficos de ventas

```python
# grafica de ventas del 2020
# sturges (k = 1 + 3.3 * log(n)) = 4.97 ~ 5
plt.hist(data["Ventas2020"], bins=5)
plt.xlabel("Ventas del 2020")
plt.ylabel("Frecuencia")
plt.title("Histograma de ventas del 2020")
```

```python
# grafica de ventas del 2021
# sturges (k = 1 + 3.3 * log(n)) = 4.97 ~ 5
plt.hist(data["Ventas2021"], bins=5)
plt.xlabel("Ventas del 2021")
plt.ylabel("Frecuencia")
plt.title("Histograma de ventas del 2021")
```

## Tabla de ventas totales

```python
# creación tabla nueva con las ventas del 2020 y 2021
#tablaventas = data[["Ventas2020", "Ventas2021"]]
tablaventas = data.loc[: , ["Ventas2020", "Ventas2021"]]
tablaventas["Suma ventas 2020-2021"] = tablaventas["Ventas2020"] + tablaventas["Ventas2021"]
tablaventas
```

```python
# suma de los valores de columnas
suma2020 = 0
suma2021 = 0
sumatotal = 0

for i in tablaventas["Ventas2020"]:
    suma2020 += i
print("La suma total del 2020 fue:", suma2020) 
for i in tablaventas["Ventas2021"]:
    suma2021 += i
print("La suma total del 2021 fue:", suma2021) 
for i in tablaventas["Suma ventas 2020-2021"]:
    sumatotal += i
print("La suma total del 2020-2021 fue:", sumatotal) 
```
