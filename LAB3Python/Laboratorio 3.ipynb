```python
# importar las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# datos csv
path = "/home/luii/Documents/Modelacionysimulacion"
file = "LAB3Python/datosincompletos2020perfiles.csv"
data = pd.read_csv(path + "/" + file)
data
```

## Creación de tabla fake

```python
# cantidad de datos
cantdatos = 25
```

```python
# a) edad
edad = np.random.uniform(20,40,cantdatos) + 4*np.random.randn(cantdatos)
```

```python
# b) años de experiencia
aniosexp = np.random.uniform(2,20,cantdatos)
```

```python
# c) años de estudio
anioestudio = np.random.uniform(2,5,cantdatos)
```

```python
fakeFrame = pd.DataFrame(
    {
        'edad':edad,
        'años de experiencia':aniosexp,
        'Años de estudio':anioestudio
    }
)

fakeFrame
```

## Fusión de tabla fake y tabla real

```python
# concatenación de las dos tablas
tablasunidas = pd.concat([data, fakeFrame], axis=1)
tablasunidas
```

## Creación de histograma para salarios max y min

```python
# Creacion de una lista para los datos en salarios

maxminsala = []

ingresosList = tablasunidas["ingresos"].values.tolist()
maxSalario = max(ingresosList)
minSalario = min(ingresosList)
maxminsala.append(max(ingresosList))
maxminsala.append(min(ingresosList))



# espacio de impresion
%matplotlib inline
plt.hist(maxminsala)
```
